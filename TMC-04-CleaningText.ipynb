```python
# Reading the raw file
import pandas as pd
pd.set_option('display.max_colwidth',100)
data = pd.read_csv("SMSSpamCollection.tsv", sep = '\t', header = None)
data.columns = ['label','body_text']
data.head()
```

### Remove punctuation

```python
import string
string.punctuation
```

```python
# creating a function to remove punctuation using list comprehension
def remove_punct(text):
        text_nopunct = "".join([char for char in text if char not in string.punctuation])
        return text_nopunct

    
#list comprehension    
data['body_text_clean'] = data['body_text'].apply(lambda x:remove_punct(x))
data.head()    
```

### Tokenization

```python
import re

# creating a function to tokenize strings

def tokenize(text):
    tokens = re.split('\W+', text) # will split on one or more non word characters
    return tokens

#list comprehension   
# as python treats capital and lower different

data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x:tokenize(x.lower()))
data.head()  
```

### Removing stop words

```python
import nltk
# nltk.corpus.stopwords.words("English")

# Function to removee stopwords using list comprehension

def remove_stopwords(text):
    text = ([word for word in text if word not in nltk.corpus.stopwords.words("English")])
    return text

#list comprehension   
# as python treats capital and lower different

data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x:remove_stopwords(x))
data.head()  

```

```python
# Stemming is the process of reducing inflected(or sometimes derived words) to their word stem or root
ps = nltk.PorterStemmer()
#dir(ps)
print(ps.stem("grows"))
print(ps.stem("grow"))
print(ps.stem("growing"))

print(ps.stem("electric")) # not a meaning
print(ps.stem("electricity"))


```

```python
# creating a function to stem words using list comprehension
def stem_text(text):
    text = [ps.stem(word) for word in text] 
    return(text)


data['body_text_stem'] = data['body_text_nostop'].apply(lambda x: stem_text(x))
data.head()
```

### Lemmatization 


```python
# Lemmatization is dictionary analysis of words to return are more better word: accuracy speed tradeoff 

wn = nltk.WordNetLemmatizer()

print(wn.lemmatize("grows"))
print(wn.lemmatize("grow"))
print(wn.lemmatize("growing"))

print(wn.lemmatize("electric")) 
print(wn.lemmatize("electricity"))

print(wn.lemmatize("goose")) 
print(wn.lemmatize("geese"))

```

```python
# creating a function to lemmatize words using list comprehension
def lemmatize_text(text):
    text = [wn.lemmatize(word) for word in text] 
    return(text)


data['body_text_lemmatize'] = data['body_text_nostop'].apply(lambda x: lemmatize_text(x))
data.head()
```

### Creating the final function for cleaning with lemmatization

```python
# A combined function for cleaning the text including lemmatization

import string
import re
import nltk

wn = nltk.WordNetLemmatizer()
ps = nltk.PorterStemmer()


def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W+', text)
    text = ([word for word in tokens if word not in nltk.corpus.stopwords.words("English")])
    text = [wn.lemmatize(word) for word in text]   
    #text = [ps.stem(word) for word in text] 
    return(text)


clean_text('abishek is a good boyy ,,goose, geese')
```
